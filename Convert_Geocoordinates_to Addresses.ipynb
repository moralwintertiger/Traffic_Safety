```python
#read in the input file:

input_file = pd.read_excel("Publicly-Owned_Properties Excel.xlsx")
```

```python
#take a quick look:

input_file.head()
```

```python
###input file looks good. Because the geopy script eats coordinates in the "lat, lon" format, build a column for that:

input_file['lat_lon'] = [str(input_file.iloc[i]['Latitude']) + "," + str(input_file.iloc[i]['Longitude']) 
                         for i in range(len(input_file['Latitude']))]
```

```python
test_coords = input_file['lat_lon'][100]
```

```python
##now build a function to return the address on each set of coordinates. Here's the base function from above:

if (not os.environ.get('PYTHONHTTPSVERIFY', '') and
getattr(ssl, '_create_unverified_context', None)):
    ssl._create_default_https_context = ssl._create_unverified_context
    #location = geolocator.reverse("34.17300524,-118.4588488")
    location = geolocator.reverse(test_coords)
```

```python
##and here's the extensible function:

def retrieve_address(input_coords):
    if (not os.environ.get('PYTHONHTTPSVERIFY', '') and
    getattr(ssl, '_create_unverified_context', None)):
        ssl._create_default_https_context = ssl._create_unverified_context
        location = geolocator.reverse(input_coords)
        return location.address
```
